## üìã 7. V√©rification des vagues (√©chelles diff√©rentes)

```python
#on va regarder la moyennes des notes donn√©es et re√ßues sur les participants pour vague 6-9
vagues_a_verifier = [6, 7, 8, 9]
colonnes_notes = [
    'pref_attractivite', 'pref_sincerite', 'pref_intelligence', 'pref_fun', 'pref_ambition', 'pref_interets',
    'pref_partenaire_attractivite', 'pref_partenaire_sincerite', 'pref_partenaire_intelligence',
    'pref_partenaire_fun', 'pref_partenaire_ambition', 'pref_partenaire_interets'
]
for vague in vagues_a_verifier:
    df_vague = df[df['vague'] == vague]
    print(f'Vague {vague} :')
    for col in colonnes_notes:
        mean_value = df_vague[col].mean()
        print(f'  Moyenne de {col} : {mean_value:.2f}')
    print('les notes sont comprises sont une distribution de 0 a 100')

```

```python
#on va controler la moyenne des notes donn√©es et re√ßues sur les participants pour vague 1-5
vagues_a_verifier = [1, 2, 3, 4, 5,10,11, 13,14,15,16,17,18,19,20,21]
colonnes_notes = [
    'pref_attractivite', 'pref_sincerite', 'pref_intelligence', 'pref_fun', 'pref_ambition', 'pref_interets',
    'pref_partenaire_attractivite', 'pref_partenaire_sincerite', 'pref_partenaire_intelligence',
    'pref_partenaire_fun', 'pref_partenaire_ambition', 'pref_partenaire_interets'
]
for vague in vagues_a_verifier:
    df_vague = df[df['vague'] == vague]
    print(f'Vague {vague} :')
    for col in colonnes_notes:
        mean_value = df_vague[col].mean()
        print(f'  Moyenne de {col} : {mean_value:.2f}')
    print('les notes sont comprises sont une distribution de 0 a 100')

```

```python
'''# on va mettre a l'echelle les notes sur une attribution de 100 points pour les vagues 6-9
# M√©thode : (valeur / somme_par_ligne) * 100

total_recues= ['attractivite_recue', 'sincerite_recue', 'intelligence_recue', 'fun_recu', 'ambition_recue', 'interets_partages_recus']
total_donnees = ['attractivite_donnee', 'sincerite_donnee', 'intelligence_donnee', 'fun_donne', 'ambition_donnee', 'interets_partages_donnes']

# Cr√©er un masque pour les vagues 6-9
mask_waves_69 = df['vague'].between(6, 9)

# Calculer les sommes par ligne UNIQUEMENT pour les vagues 6-9
somme_recues = df.loc[mask_waves_69, total_recues].sum(axis=1)
somme_donnees = df.loc[mask_waves_69, total_donnees].sum(axis=1)

# Normaliser chaque colonne
for col in total_recues:
    mask_valid = (somme_recues > 0) & df.loc[mask_waves_69, col].notna()
    df.loc[mask_waves_69 & mask_valid, col] = (
        df.loc[mask_waves_69 & mask_valid, col] / somme_recues[mask_valid] * 100
    )

for col in total_donnees:
    mask_valid = (somme_donnees > 0) & df.loc[mask_waves_69, col].notna()
    df.loc[mask_waves_69 & mask_valid, col] = (
        df.loc[mask_waves_69 & mask_valid, col] / somme_donnees[mask_valid] * 100
    )'''
```
